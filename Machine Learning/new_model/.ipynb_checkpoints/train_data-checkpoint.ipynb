Import Library in Here

```python

from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

Load Job Dataset

```python
import pandas as pd
job = pd.read_csv("job.csv", index_col=0)
job.head(10)
```

Drop some columns because it doesn't matter in our training

```python
job.drop(columns=["Employment", "Gender", "MainBranch", "Country"], axis=1, inplace=True)
job.head(10)
```

Separate between feature X and feature Y

```python
df_feature_x = pd.DataFrame(
    [job["Age"], job["Accessibility"], job["EdLevel"], job["MentalHealth"], job["YearsCode"], job["YearsCodePro"], 
     job["PreviousSalary"], job["HaveWorkedWith"], job["ComputerSkills"]]).transpose()
df_feature_x.head(10)
```

```python
df_feature_y = pd.DataFrame(job["Employed"])
df_feature_y.head(10)
```

One-hot Encode some Columns

```python
df_onehot_age = pd.get_dummies(df_feature_x["Age"], prefix="Age")
df_onehot_access = pd.get_dummies(df_feature_x["Accessibility"], prefix="Accessibility")
df_onehot_edlevel = pd.get_dummies(df_feature_x["EdLevel"], prefix="EdLevel")
df_onehot_mental = pd.get_dummies(df_feature_x["MentalHealth"], prefix="MentalHealth")
df_feature_x = pd.concat([df_onehot_age, df_onehot_access, df_onehot_edlevel, df_onehot_mental, df_feature_x], axis=1)
df_feature_x.drop(columns=["Age", "Accessibility", "EdLevel", "MentalHealth"], inplace=True)
df_feature_x.head(10)
```

Make it so when True=1, False=0

```python
df_feature_x["Age_<35"] = df_feature_x["Age_<35"].astype(int)
df_feature_x["Age_>35"] = df_feature_x["Age_<35"].astype(int)
df_feature_x["Accessibility_No"] = df_feature_x["Accessibility_No"].astype(int)
df_feature_x["Accessibility_Yes"] = df_feature_x["Accessibility_Yes"].astype(int)
df_feature_x["EdLevel_Master"] = df_feature_x["EdLevel_Master"].astype(int)
df_feature_x["EdLevel_NoHigherEd"] = df_feature_x["EdLevel_NoHigherEd"].astype(int)
df_feature_x["EdLevel_Other"] = df_feature_x["EdLevel_Other"].astype(int)
df_feature_x["EdLevel_PhD"] = df_feature_x["EdLevel_PhD"].astype(int)
df_feature_x["EdLevel_Undergraduate"] = df_feature_x["EdLevel_Undergraduate"].astype(int)
df_feature_x["MentalHealth_No"] = df_feature_x["MentalHealth_No"].astype(int)
df_feature_x["MentalHealth_Yes"] = df_feature_x["MentalHealth_Yes"].astype(int)
df_feature_x.head(10)
```

```python
df_feature_y["Employed"] = df_feature_y["Employed"].astype(float)
df_feature_y.head(10)
```

Standardizing YearsCode, YearsCodePro, PreviousSalary, and ComputerSkill

```python
norm_scaler = StandardScaler()

years_code = df_feature_x["YearsCode"].values.reshape(-1, 1)
years_code_pro = df_feature_x["YearsCodePro"].values.reshape(-1, 1)
prev_salary = df_feature_x["PreviousSalary"].values.reshape(-1, 1)
comp_skill = df_feature_x["ComputerSkills"].values.reshape(-1, 1)

new_values_years_code = norm_scaler.fit_transform(years_code)
new_values_years_pro = norm_scaler.fit_transform(years_code_pro)
new_values_salary = norm_scaler.fit_transform(prev_salary)
new_values_skill = norm_scaler.fit_transform(comp_skill)

df_feature_x["YearsCode"] = new_values_years_code
df_feature_x["YearsCodePro"] = new_values_years_pro
df_feature_x["PreviousSalary"] = new_values_salary
df_feature_x["ComputerSkills"] = new_values_skill

df_feature_x.head(10)
```

MinMaxing YearsCode, YearsCodePro, PreviousSalary, and ComputerSkill between -1 and 1

```python
scaler = MinMaxScaler(feature_range=(-1, 1))

new_values_years_code = scaler.fit_transform(new_values_years_code)
new_values_years_pro = scaler.fit_transform(new_values_years_pro)
new_values_salary = scaler.fit_transform(new_values_salary)
new_values_skill = scaler.fit_transform(new_values_skill)

df_feature_x["YearsCode"] = new_values_years_code
df_feature_x["YearsCodePro"] = new_values_years_pro
df_feature_x["PreviousSalary"] = new_values_salary
df_feature_x["ComputerSkills"] = new_values_skill

df_feature_x.head(10)
```

Terpaksa harus di drop dulu yang HaveWorkedWith

```python
df_feature_x.drop(columns=["HaveWorkedWith"], inplace=True)
df_feature_x.head(10)
```

Split data into train, validation, and test

```python
from sklearn.model_selection import train_test_split
X_train, X_, Y_train, Y_ = train_test_split(df_feature_x, df_feature_y, test_size=0.4, random_state=1)
X_valid, X_test, Y_valid, Y_test = train_test_split(X_, Y_, test_size=0.5, random_state=1)

print(f"Data before splitting: {len(df_feature_x)}")
print(f"Train data: {len(X_train)}")
print(f"Validation data: {len(X_valid)}")
print(f"Test data: {len(X_test)}")
```

Try to train the model using TensorFlow (Neural Network)

```python
import tensorflow as tf
len_feature = len(df_feature_x.columns)

model = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(len_feature,)),
    tf.keras.layers.Dense(128, activation="relu"),
    tf.keras.layers.Dense(64, activation="relu"),
    tf.keras.layers.Dense(32, activation="relu"),
    tf.keras.layers.Dense(1, activation="sigmoid")
])

model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3), 
              loss=tf.keras.losses.BinaryCrossentropy(),
              metrics=["accuracy"])

history = model.fit(X_train.astype("float32"), Y_train.astype("float32"), epochs=100, validation_data=(X_valid.astype("float32"), Y_valid.astype("float32")))
```
