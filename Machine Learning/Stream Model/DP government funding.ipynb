```python
import pandas as pd
import numpy as np
```

```python
filename = "cleaned_jobactive.zip"
df = pd.read_csv(filename, compression="zip")
df.head(5)
```

```python
for col in df.columns:
    print(col + " : ")
    print(df[col].unique())
    print()
```

```python
col_fill={'CALD':0, 'INDIGENOUS':0, 'HOMELESS':0, 'PWD':0, 'REFUGEE':0, 'EX_OFFENDER':0}
df.fillna(col_fill,inplace=True)
df.head()
```

```python
col_drop=['year', 'month', 'year_month', 'TRANSACTION_ID', 'DATA_AS_AT', 'STATE',"ACTIVITY_TYPE_DESC", 
          "ACTIVITY_SUB_DESC", "DERIVED_DATE_OF_SERVICE", "EDUCATION_LEVEL_DESC"]
df.drop(col_drop, inplace=True, axis=1)
df.dropna(axis=0, inplace=True)
col_drop2=["EF_CATEGORY_DESCRIPTION", "EF_CATEGORY_ADJ", "EXPENDITURE"]
df.drop(col_drop2, inplace=True, axis=1)
df=df.astype({'INDIGENOUS':'int', 'HOMELESS':'int', 'PWD':'int', 'REFUGEE':'int', 'EX_OFFENDER':'int', 'CALD':'int'})
```

```python
df.drop_duplicates(inplace=True)
df.head()
```

```python
df.drop(df[df['STREAM_PLACEMENT_DESC']=="Stream Volunteer"].index, inplace=True)
```

```python
df.to_csv("stream_classify.csv")
```
